```python
import pandas as pd
import csv
df = pd.read_csv('Indonesian_Manually_Tagged_Corpus_ID.tsv', sep='\t', header=None, names=['word','tag'], converters={'word': lambda x: x.lower()}, quoting=csv.QUOTE_NONE)
split_index_pertama = int(df[df['word'].str.contains('id=1001')].index[0])
split_index_kedua = int(df[df['word'].str.contains('id=1021')].index[0])
# dataTest = df[:200]
#mengetahui nilai indeks terhadap data yang ada di corpus bahasa indonesia 
variabel1 = split_index_pertama
print(split_index_pertama)
print(split_index_kedua)
# dataTest
```

```python
df1 = pd.DataFrame() #df1 merupakan variable yang akan menyimpan data ke dalam bentuk dataframe
df1
```

```python
for i in range(split_index_pertama):
    df1 = df1.append(df.loc[i])
df1 #menyimpan data f1 yakni dari indeks 1 ke 29177 ke dalam data frame
```

```python
df1 = df1.dropna()
df1 #menghilangkan post tag yang 'NaN'
```

```python
df_train = df1[:split_index_pertama].reset_index(drop=True) #melakukan reset pada index dari 0
# df_train
```

```python
split_index_pertama
```

```python
df2 = pd.DataFrame() #membuat variabel df sebagai dataframe untuk menyimpan data uji
# i = split_index_pertama
#memasukkan 20 kalimat setelah 1000 kalimat pertama ke dalam df2 yang berarti merupakan data uji
for i in range(variabel1, split_index_kedua): 
    df2 = df2.append(df.loc[i])
# df2
```

```python
df2 = df2.dropna() #menghilangkan post tag yang 'NaN' 
# df2
```

```python
df_test = df2.reset_index(drop=True) #melakukan reset pada index dari 0
df_test
```

```python
# df.groupby(['word','tag']).size()
#menampilkan tabel yang berisikan kata beserta post tag nya dalam bentuk dataframe terhadap kata latih
tabel_train = df_train.groupby(['word','tag']).size().unstack(fill_value=0)
```

```python
tabel_train
```

```python
tabel_test = df_test.groupby(['word','tag']).size().unstack(fill_value=0)
```

```python
tabel_test #menampilkan tabel yang berisikan kata beserta post tag nya dalam bentuk dataframe terhadap kata uji
```

```python
# maks_tabel_train = pd.DataFrame()
maks_tabel_train = tabel_train.idxmax(axis=1) #mengetahui nilai pos tag yang sering muncul terhadap masing masing kata pada data latih
# maks_tabel_train
data_dict_train = maks_tabel_train.to_dict()
data_dict_train
```

```python
#membandingkan data uji dan data latih untuk dapat mengetahui nilai keakuratan terhadap baseline yang didapat
salah = [] #menyimpan kata dengan post tag yang salah
# kata_test = data_dict_test.keys()
# tag_test = data_dict_test.values()
# kata_train = data_dict_train.keys()
# tag_train = data_dict_train.values()

for k, v in data_dict_test.items():
#     print('uji: ', k, v)
    simpan_key = k
    simpan_value = v
    for a, b in data_dict_train.items():
#         print('latih: ', a, b)
        if (simpan_key == a):
#             print('latih: ', a, b)
            if (simpan_value != b):
                print('salah:', simpan_key, simpan_value, 'seharusnya:', a , b)
                postag_salah = a,b
                salah.append(postag_salah)
    
```
